# Import libraries

```python
from numpy import array
from numpy import random
from numpy import dot
from scipy.special import softmax
```

**Define a text sequence and do manual encoding**
>Tom bit the dog

These word embeddings normally would have been generated by an encoder; however, for this particular example, you will define them manually

```python
word=["Tom", "bit","the", "dog"]
wordEn=[]
```

```python
# encoder representations of four different words
wordEn.append(array([1, 0, 0])) #Tom
wordEn.append([0, 1, 0])        #bit
wordEn.append(array([1, 1, 0])) #the
wordEn.append(array([0, 0, 1])) #dog
for i in range(len(word)):
  print(word[i], wordEn[i])
```

**Prepare the encoded text sequence for the encoder**

```python
# stacking the word embeddings into a single array
words = array([wordEn[0], wordEn[1], wordEn[2], wordEn[3]])
print(words)
```

**Generate weights for Q, K, V**<br>
The number of rows of each of the matrices is equal to the dimensionality of the word embeddings (which in this case is three) to allow us to perform the matrix multiplication

```python
# generating the weight matrices
random.seed(42)
W_Q = random.randint(3, size=(3, 3))
W_K = random.randint(3, size=(3, 3))
W_V = random.randint(3, size=(3, 3))
print("W_Q=", W_Q)
print("W_K=", W_K)
print("W_V=", W_V)
```

The query, key, and value vectors for each word are generated by multiplying each word embedding by each of the weight matrices

```python
q=[]
k=[]
v=[]
# generating the queries, keys and values
q.append(words[0] @ W_Q)
k.append(words[0] @ W_K)
v.append(words[0] @ W_V)

q.append(words[1] @ W_Q)
k.append(words[1] @ W_K)
v.append(words[1] @ W_V)

q.append(words[2] @ W_Q)
k.append(words[2] @ W_K)
v.append(words[2] @ W_V)

q.append(words[3] @ W_Q)
k.append(words[3] @ W_K)
v.append(words[3] @ W_V)

print("q= ",q)
print("k= ",k)
print("v= ",v)
```

```python
# generating the queries, keys and values
Q = words @ W_Q
K = words @ W_K
V = words @ W_V
print("Q= ", Q)
print("K= ", K)
print("V= ", V)
```

**Score the query vectors**

```python
KT = K.transpose()
print(Q)
print()
print(KT)

```

```python
# scoring the query vectors against all key vectors
scores = Q @ KT
scores
```

**Calculations for first row of scores**
[202] * [222] = 4+0+4=8<br>
[202] * [021] = 0+0+2=2<br>
[202] * [243] = 4+0+6=10<br>
[202] * [011] = 0+0+2=2<br>

**Softmax**<br>
The softmax function transforms each element of a collection by<br>
>computing the exponential of each element divided by the sum of the exponentials of all the elements.

```python
# computing the weights by a softmax operation
weights = softmax(scores / K.shape[1] ** 0.5, axis=1)
weights
```

Softmax -- column sum = 1.0<br>

```python
weights.sum(axis=1)
```

Recall the dimensions are values that give meaning to the words.

```python
# computing the attention by a weighted sum of the value vectors
attention = weights @ V

print("word     dim 1    dim 2      dim3")
for i in range(len(word)):
  print(word[i],attention[i])
```
